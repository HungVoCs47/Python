```python
# Permutations
# generated algorithm
def permute(A):
    if(len(A) == 1):
        return [tuple(A)]
    permutations = []
    for x in A:
        for y in permute(A-{x}):
            permutations.append((x,)+y)
    return permutations
```

```python
A = {1, 2, 3}
```

```python
permute_all = set(permute(A))
print("Permutations of {}: {}".format(A,permute_all))
print("Number of permutations: ", len(permute_all))
```

```python
permute_all = set(itertools.permutations(A))
print("Permutations of {}: {}".format(A,permute_all))
print("Number of permutations: ", len(permute_all))
```

```python
# Partial Permutations
def pp(A,k):
    if(k==1):
        return [(x,) for x in A]
    permutations = []
    for x in A:
        for y in pp(A-{x}, k-1):
            permutations.append((x,)+y)
    return permutations
        
```

```python
A = {1, 2, 3, 4}
k = 3
n = len(A)
```

```python
per = pp(A, k)
print("{}-permutations of {}: {}".format(k,A,per))
print(len(per))
```

```python
# Combinations
def com(A,k):
    if(k == 1):
        return [{x} for x in A]
    com_ = []
    #print(com_)
    for x in A:
        for y in com(A-{x}, k-1):
            #print(com_)
            if {x}.union(y) not in com_:
                com_.append({x}.union(y))
    return com_
```

```python
A = {1, 2, 3, 4, 5}
k = 3
```

```python
per = com(A, k)
print("{}-permutations of {}: {}".format(k,A,per))
print(len(per))
```

```python
choose_k = list(itertools.combinations(A,k))
print("{}-combinations of {}: {}".format(k,A,choose_k))
print("Number of combinations = {}".format(len(choose_k)))
```
