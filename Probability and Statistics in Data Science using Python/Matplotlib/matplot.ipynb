```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
# Plot y = cos(x)
x = np.arange(0, 4 * np.pi, 0.1)
y_cos = np.cos(x)

plt.figure()
plt.plot(x, y_cos)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
x = np.arange(0, 10, 1)
y_1 = 2**x
y_2 = x**2
plt.plot(x, y_1, label = '$2^x$', color='g', linestyle='--', marker='s')
plt.plot(x, y_2, label = '$x^2$', color='r', linestyle='-', marker='o')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# Data
x = np.arange(0, 10, 1) # x in [0, 10) 
y_1 = 2 ** x
y_2 = x ** 2

plt.figure()
# Specifying color, linestyle and marker using positional arguments
plt.plot(x, y_1, 'g--s', label = '$2^x$')
plt.plot(x, y_2, 'r-o', label = '$x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')
plt.show()
```

```python
x = np.arange(-10, 10, 0.1)
y = x ** 3

plt.figure()
plt.plot(x, y, label = '$x^3$')
plt.xlabel('$x$', fontsize = 12) # The fontsize can be set here as well
plt.ylabel('$y$', fontsize = 12)
plt.title('$y = x^3$', fontsize = 16) # Set title and its fontsize
plt.legend(loc = 'upper left')
plt.grid()
plt.show()
```

```python
# suplots
x = np.arange(0, 6*np.pi, 0.2)
y_1 = np.cos(x)
y_2 = np.sin(2*x)


# plot y = cos(x)
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, y_1, label = '$cos(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')



plt.subplot(2, 1, 2)
plt.plot(x, y_2, label = '$\sin(2x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'lower left')
plt.show()
```

```python
# Shared Axes
x = np.arange(0, 6 * np.pi, 0.2)
y_1 = np.cos(x)
y_2 = np.sin(2 * x)
y_3 = y_1 + y_2

plt.figure()

plt.subplot(3, 1, 1)
plt.plot(x, y_1, label = '$cos(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')


plt.subplot(3, 1, 2)
plt.plot(x, y_2, label = '$sin(2x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')

plt.subplot(3, 1, 3)
plt.plot(x, y_3, label = '$sin(2x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')
```

```python
x = np.arange(0, 7, 1) # x in [2, 7) 
y = x
 
plt.figure()
plt.bar(x, y, label = '$x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# Multiple Axis
calories = [380.70, 420.98, 454.91, 406.45, 446.16, 498.08, 504.54, 459.05, 459.55, 484.79]
countries = ['India', 'Japan', 'Korea', 'China', 'Thai', 'Italy', 'France', 'Greece', 'Mexico', 'US']
obesity_rates = [3.9, 4.3, 4.7, 6.2, 10, 19.9, 21.6, 24.9, 28.9, 36.2]
```

```python
fig, ax1 = plt.subplots(figsize = (8, 5))
ax1.bar(countries, obesity_rates, color='C8')
ax1.set_ylabel('obesity rate(%)', color='C8')

ax2 = ax1.twinx()
ax2.plot(countries, calories, color='C0')
ax2.set_ylabel('calories', color='C0')


plt.show()
```

```python
# Scatter Plot
# Data
x = np.arange(0, 10, 0.5) # x in [0, 10) 
print(len(x))
noise = np.random.randn(len(x)) # Generate standard normal random variables
print(noise)
y = 2 * x + 3 + noise

plt.figure()
plt.scatter(x, y)
plt.plot(x, 2 * x + 3, color = 'r')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
